```python
import os
from spotipy.oauth2 import SpotifyOAuth
import spotipy
from dotenv import load_dotenv

load_dotenv()

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id=os.getenv("SPOTIPY_CLIENT_ID"),
    client_secret=os.getenv("SPOTIPY_CLIENT_SECRET"),
    redirect_uri=os.getenv("SPOTIPY_REDIRECT_URI"),
    scope="user-top-read user-read-recently-played playlist-read-private"
))

sp.current_user()
```

```python
recent_tracks = sp.current_user_recently_played(limit=10)

for item in recent_tracks["items"]:
    track = item["track"]
    print(f"{track['name']} - {track['artists'][0]['name']}")
```

```python
top_artists = sp.current_user_top_artists(limit=10, time_range="medium_term")

for artist in top_artists["items"]:
    print(artist["name"])
```

```python
top_tracks = sp.current_user_top_tracks(limit=10, time_range="medium_term")

for track in top_tracks["items"]:
    print(f"{track['name']} - {track['artists'][0]['name']}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
df1["danceability"].hist(bins=30)
plt.title("Danceability Distribution")
plt.xlabel("Danceability")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
df1["energy"].hist(bins=30)
plt.title("Energy Distribution")
plt.xlabel("Energy")
plt.ylabel("Count")
plt.show()
```

```python
genre_stats = (
    df1
    .groupby("track_genre")[["energy", "danceability", "valence"]]
    .mean()
    .sort_values(by="energy", ascending=False)
)

genre_stats.head(10)
```

```python
top_energy_genres = genre_stats.head(10)

plt.figure()
top_energy_genres["energy"].plot(kind="bar")
plt.title("Top 10 Most Energetic Genres")
plt.xlabel("Genre")
plt.ylabel("Average Energy")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
df1["track_name_clean"] = (
    df1["track_name"]
    .str.lower()
    .str.replace(r"\(.*\)", "", regex=True)
    .str.replace(r"-.*", "", regex=True)
    .str.strip()
)
```

```python
spotify_df["track_name_clean"] = (
    spotify_df["track_name"]
    .str.lower()
    .str.replace(r"\(.*\)", "", regex=True)
    .str.replace(r"-.*", "", regex=True)
    .str.strip()
)
```

```python
final_df = df1.merge(
    spotify_df,
    on="track_name_clean",
    how="inner"
)
```
